## Orthogonal Projection  

Orthogonal projection is a method of projecting a vector onto another vector or subspace such that the resulting vector lies entirely within the subspace and is perpendicular to the orthogonal component. It minimizes the distance between the original vector and the subspace.  

---

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
p_vec = np.array([3, 5])  # Projection vector
b_vec = np.array([0.5, 3])  # Base vector

plt.figure(figsize=(8, 8))
plt.arrow(0, 0, p_vec[0], p_vec[1], head_width=0.2, length_includes_head=True, color='red', label='Projection Vector (p_vec)')
plt.arrow(0, 0, b_vec[0], b_vec[1], head_width=0.2, length_includes_head=True, color='green', label='Base Vector (b_vec)')

# Configure plot
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Projection Vector and Base Vector')
plt.xlim(-1, 4)
plt.ylim(-1, 6)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python

# Compute projection
beta = np.dot(b_vec, p_vec) / np.dot(p_vec, p_vec)
parallel_vec = beta * p_vec  # Projection of b_vec onto p_vec
project_vec = b_vec - parallel_vec  # Orthogonal component

# Plot the vectors
plt.figure(figsize=(8, 8))
plt.arrow(0, 0, p_vec[0], p_vec[1], head_width=0.2, length_includes_head=True, color='red', label='Projection Vector (p_vec)')
plt.arrow(0, 0, b_vec[0], b_vec[1], head_width=0.2, length_includes_head=True, color='green', label='Base Vector (b_vec)')
plt.arrow(0, 0, parallel_vec[0], parallel_vec[1], head_width=0.2, length_includes_head=True, color='yellow', label='Projected Vector')
plt.arrow(parallel_vec[0], parallel_vec[1], project_vec[0], project_vec[1], head_width=0.2, length_includes_head=True, color='blue', label='Orthogonal Component')

# Add dashed line for orthogonal projection
plt.plot([b_vec[0], parallel_vec[0]], [b_vec[1], parallel_vec[1]], 'k--', label='Orthogonal Line')

# Configure plot
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Orthogonal Projection')
plt.xlim(-1, 4)
plt.ylim(-1, 6)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
